```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('OnlineRetail.csv', encoding= 'latin-1')

df
```

```python
# Data types

df.info()
```

```python
# Change Data type of InvoiceNo, CustomerID & Invoice date

df = df.astype({'CustomerID' : 'string', 
                'InvoiceNo' : 'string'})

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df.dtypes
```

```python
df.isna().sum()
```

```python
# Check if quantity is 0

df[df['Quantity']<=0]
```

```python
df1 = df[(df['CustomerID'].isna() == False) & (df['Quantity'] > 0) & (df['InvoiceNo'].isna() == False)].reset_index(drop = True)

df1

```

```python
df1.isna().sum()
```

```python
# Remove leading annd trailing spaces in 'description'

df1['Description'] = df1['Description'].str.strip()

df1
```

```python
# Unique invoice # based on country

df2 = df1.groupby('Country')['InvoiceNo'].nunique().sort_values(ascending = False)

df2
```

```python
# United Kingdom

uk = df1[df1['Country'] == 'United Kingdom']

uk
```

```python
uk_purchase = uk.groupby(['InvoiceNo', 'Description'])['Quantity'].count().unstack().fillna(0)

uk_purchase
```

```python
# pip --version
```

```python
# One-hot encoding

def one_hot_encode(x) :
    
     if x <= 0 :
        return 0
     if x >= 1 : 
        return 1
    
uk_purchase_encode = uk_purchase.applymap(one_hot_encode)    

uk_purchase_encode
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
freq_items = apriori( uk_purchase_encode.astype('bool'), min_support = 0.02, use_colnames = True)

freq_items
```

```python
# Association rule mining
rules = association_rules(freq_items, metric='lift', min_threshold =1)

rules
```

```python
rules = rules[rules['confidence'] >= 0.5]

rules.sort_values(['confidence', 'lift'], ascending= False).reset_index(drop = True).head(10)
```
